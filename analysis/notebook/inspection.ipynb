```python
import pandas as pd
import numpy as np

base_dir = "../../data/lab/"
stg = "upoker-p_nlh2_30bb_cfrs_r0_48300000"
state = "open"
file = stg + "_" + state
df = pd.read_csv(base_dir+file+".csv")
df.head()

rank = "AKQJT98765432"

df_collection = {}
for action in df.columns[7:]:
    idx = df.columns.get_loc(action)
    df_collection[idx] = pd.DataFrame(0, index=["A","K","Q","J","T","9","8","7","6","5","4","3","2"],
                columns=["A","K","Q","J","T","9","8","7","6","5","4","3","2"])

df["card1"] = df["card1"].str[0]
df["card2"] = df["card2"].str[0]
    
for c1 in rank:
    for c2 in rank:
        if c1 != c2:
            suited = df[(df["card1"]==c1) & (df["card2"]==c2) & (df["suited"]==1)].values
            unsuited = df[(df["card1"]==c1) & (df["card2"]==c2) & (df["suited"]==0)].values
            if(len(suited)):
                for key in df_collection.keys():
                    df_collection[key].iloc[rank.find(c1), rank.find(c2)] = suited[0][key]
            if(len(unsuited)):
                for key in df_collection.keys():
                    df_collection[key].iloc[rank.find(c2), rank.find(c1)] = unsuited[0][key]
        else:
            for key in df_collection.keys():
                df_collection[key].iloc[rank.find(c1), rank.find(c2)] = df[(df["card1"]==c1) & (df["card2"]==c2)].values[0][key]

print(df_collection.keys())
print(df.columns[7:])

import seaborn as sb
import math
import matplotlib.pyplot as plt

plot_rows = math.floor(len(df_collection.keys())/2)+1
plot_cols = 2
print(plot_rows, plot_cols)
fig, ax = plt.subplots(plot_rows, plot_cols, constrained_layout=True, sharey=True, figsize=(15,20))


for idx,key in enumerate(df_collection.keys()):
    row = math.floor(idx/2)
    col = math.floor(idx%2)
    ax[row][col].set_title(file + " | " + df.columns[key])
    sb.heatmap(df_collection[key], annot=True, cmap="PuRd", ax=ax[row][col], linewidths=.5, vmin=0, vmax=1)
```
